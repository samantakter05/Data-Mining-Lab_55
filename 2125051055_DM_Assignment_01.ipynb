```python
ratings = pd.read_csv('/content/drive/MyDrive/DM Lab Project 01/ratings.csv')

user_movie_matrix = ratings.pivot(index='userId', columns='movieId', values='rating')

movie_similarity = user_movie_matrix.corr(method='pearson')

movie_similarity
```

```python
ratings = pd.read_csv("/content/drive/MyDrive/DM Lab Project 01/ratings.csv")
movies = pd.read_csv("/content/drive/MyDrive/DM Lab Project 01/movies.csv")

def recommend_movies(movie_id, top_n=5):

    if movie_id not in movie_similarity:
        return "Movie not found."

    similar_movies = movie_similarity[movie_id].dropna().sort_values(ascending=False)[1:top_n+1]

    recommended = movies[movies["movieId"].isin(similar_movies.index)][["movieId", "title"]]

    return recommended

recommendations = recommend_movies(1, top_n=5)
recommendations
```

```python
def recommend_movies_with_highest_avg_rating(user_id, top_n=5):

    user_ratings = ratings[ratings["userId"] == user_id]

    rated_movies = set(user_ratings["movieId"])

    all_movies = set(movies["movieId"])

    unrated_movies = all_movies - rated_movies

    unrated_movies_details = movies[movies["movieId"].isin(unrated_movies)]

    avg_ratings = ratings.groupby("movieId")["rating"].mean()

    recommended_movies = unrated_movies_details.merge(
        avg_ratings, on="movieId", how="left"
    ).sort_values(by="rating", ascending=False).head(top_n)

    return recommended_movies[["movieId", "title", "rating"]]


recommended_movies = recommend_movies_with_highest_avg_rating(user_id=55, top_n=5)
recommended_movies
```
